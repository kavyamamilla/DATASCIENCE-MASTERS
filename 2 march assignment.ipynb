### Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy. As such, it offers a viable open source alternative to MATLAB. Developers can also use matplotlib's APIs (Application Programming Interfaces) to embed plots in GUI applications.

### Bar Chart, Scatter Plot, Histograms, Contour Plots, Box Plot, Pie Chart are using matplotlib.

### Scatter plots are the graphs that present the relationship between two variables in a data-set. 

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.scatter(x,y , c = '#223BDC',alpha=.9,marker='X')
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
```

### subplot() function adds subplot to a current figure at the specified grid position.  It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. 

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y, marker="x", color="green")
plt.subplot()
x=np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y, marker="x", color="red")
plt.subplot()
x=np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y, marker="x", color="black")
plt.subplot()
np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y, marker="x", color="indigo")
plt.subplot()
```

### A barplot (or barchart) is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

###  The bar graph is used to compare the items between different groups over time. Bar graphs are used to measure the changes over a period of time.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(5,2))
plt.bar(company,profit,color = 'orange')
plt.xlabel("company")
plt.ylabel("profit")
plt.title("company vs profit")
plt.grid()
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(5,2))
plt.barh(company,profit,color = 'orange')
plt.xlabel("company")
plt.ylabel("profit")
plt.title("company vs profit")
plt.grid()
```
